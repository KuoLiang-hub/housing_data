```python
import pandas as pd
import plotly.graph_objects as go
```

```python
# read data and set Data from text to datetime object
nj_data = pd.read_csv('./raw_data/NJ-2024-05.csv')
nj_data['Date'] = pd.to_datetime(nj_data['Date'])
```

```python
# lis of south and north jersey
north_jersey = ["Bergen County","Middlesex County","Essex County","Hudson County","Monmouth County","Union County","Passaic County","Morris County", "Mercer County","Somerset County","Sussex County","Hunterdon County","Warren County"]
south_jersey = ["Ocean County","Camden County","Burlington County", "Gloucester County","Atlantic County","Cumberland County","Cape May County","Salem County"]
```

```python
# TODO add income level to it too, maybe a second y; add home to income ratio.
df_income = pd.read_csv("./raw_data/atlantic_ct_household_inc.csv")
df_income['Date'] = pd.to_datetime(df_income['Date'])
df_interest = pd.read_csv("./raw_data/mort_rate_by_year.csv")
df_interest['Date'] = pd.to_datetime(df_interest['Date'])

df_merged = pd.merge_asof(nj_data, df_income, on='Date', direction='backward')
df_merged = pd.merge_asof(df_merged, df_interest, on='Date', direction='backward')

def calRation(row):
    return None if row['Atlantic County'] is None or row['Median_Household_Income_Atlantic_County'] is None else row['Atlantic County']/row['Median_Household_Income_Atlantic_County']
df_merged['Home_to_income_ratio'] = df_merged.apply(calRation, axis=1)
# df_merged.head()
```

```python
df_merged.to_csv("./df_mge.csv")
```

```python
# TODO: try use plotly for this too...
fig = go.Figure()

fig.add_trace(go.Scatter(x=df_merged['Date'], y=df_merged['Atlantic County'], name='Atlantic County Median Home Value', yaxis='y'))
fig.add_trace(go.Scatter(x=df_merged['Date'], y=df_merged['Home_to_income_ratio'], name='Home_to_income_ratio', yaxis='y2'))
fig.add_trace(go.Scatter(x=df_merged['Date'], y=df_merged['30-year fixed-rate average'], name='30-year fixed-rate average', yaxis='y3'))

# Create axis objects
fig.update_layout(
    xaxis=dict(domain=[0.3, 0.9]),
				# create 1st y axis
		yaxis=dict(
			title="Median Home Value (Atlantic County)",
			titlefont=dict(color="#1f77b4"),
			tickfont=dict(color="#1f77b4")
        ),

		# create 2nd y axis
		yaxis2=dict(
            title="Home Value to Median Income Ration (Atlantic County)", overlaying="y",
			side="left", 
            position=0.15
        ),
		# create 3rd y axis
		yaxis3=dict(
			title="mortgage rate",
			# anchor="x", 
            overlaying="y", 
            side="left",
			position=0.99
		)

)


# title
fig.update_layout(
	title_text="Home value and income ratio comparison",
	width=1200,
)

fig.show()
```
